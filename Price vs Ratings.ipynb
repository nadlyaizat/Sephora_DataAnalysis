### Aim : Investigate the relationship between product ratings and price for skincare products.

### Hypothesis: Skincare products with higher prices tend to have higher average ratings compared to lower-priced products.

#### Step 1: Import necessary libraries and load the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/Users/nadlyaizatnudri/Documents/Data Analysis Practice Kaggle/Sephora /archive/product_info.csv')
```

#### Step 2: Perform data exploration and preprocessing.

```python
# Check the structure and summary statistics of the dataset
print(df.head())
```

```python
print(df.info())
```

#### Step 3: Filter the dataset to include only skincare products.

```python
skincare_df = df[df['primary_category'] == 'Skincare']
```

#### Step 4: Visualize the distribution of ratings and prices.

```python
# Create a scatter plot of ratings vs. prices
plt.scatter(skincare_df['rating'], skincare_df['price_usd'])
plt.xlabel('Rating')
plt.ylabel('Price (USD)')
plt.title('Relationship between Ratings and Prices for Skincare Products')
plt.show()

# Calculate the correlation between ratings and price
correlation = skincare_df['price_usd'].corr(skincare_df['rating'])
print('Correlation between Ratings and Price:', correlation)

```

#### Step 5: Calculate the average rating for different price ranges.

```python
# Create price ranges for grouping
price_ranges = pd.cut(skincare_df['price_usd'], bins=[0, 25, 50, 100, 200, float('inf')], labels=['<25', '25-50', '50-100', '100-200', '200+'])

# Calculate the average rating for each price range
avg_rating_by_price = skincare_df.groupby(price_ranges)['rating'].mean()
print(avg_rating_by_price)
```

#### Step 6: Visualize the average ratings for different price ranges.

```python
# Create a bar plot of average ratings by price range
plt.bar(avg_rating_by_price.index, avg_rating_by_price.values)
plt.xlabel('Price Range (USD)')
plt.ylabel('Average Rating')
plt.title('Average Ratings for Different Price Ranges')
plt.show()
```
